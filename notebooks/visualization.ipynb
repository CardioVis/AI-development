```python
import numpy as np
import matplotlib.pyplot as plt

crf0 = 23
crf = np.arange(16, 33)
d = crf - crf0

qstep_ratio = 2 ** (d / 6)
mse_ratio = 2 ** (d / 3)
psnr_delta = -10 * np.log10(mse_ratio)

fig, ax = plt.subplots(1, 3, figsize=(15, 4))

ax[0].plot(crf, qstep_ratio, marker='o')
ax[0].axvline(crf0, ls='--', c='gray')
ax[0].set_title('Qstep ratio vs CRF (baseline CRF=23)')
ax[0].set_xlabel('CRF'); ax[0].set_ylabel('Qstep / Qstep23'); ax[0].grid(True)

ax[1].plot(crf, mse_ratio, marker='o', color='tomato')
ax[1].axvline(crf0, ls='--', c='gray')
ax[1].set_title('MSE ratio vs CRF')
ax[1].set_xlabel('CRF'); ax[1].set_ylabel('MSE / MSE23'); ax[1].grid(True)

ax[2].plot(crf, psnr_delta, marker='o', color='seagreen')
ax[2].axvline(crf0, ls='--', c='gray')
ax[2].axhline(0, ls='--', c='gray')
ax[2].set_title('ΔPSNR vs CRF')
ax[2].set_xlabel('CRF'); ax[2].set_ylabel('dB vs CRF23'); ax[2].grid(True)

plt.tight_layout()
plt.show()
```
